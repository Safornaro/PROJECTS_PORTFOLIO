## Analisi RFM (Recency, Frequency, Monetary)

```python
oggi = pd.to_datetime("2025-08-06")


df["Recency"] = (oggi - df["LastTransactionDate"]).dt.days
```

```python
df.head(5)
```

```python
df["R_Score"] = None

for i in df.index:
    if df.loc[i, "Recency"] <= 30:
        df.loc[i, "R_Score"] = 5
    elif df.loc[i, "Recency"] <= 90:
        df.loc[i, "R_Score"] = 4
    elif df.loc[i, "Recency"] <= 180:
        df.loc[i, "R_Score"] = 3
    elif df.loc[i, "Recency"] <= 365:
        df.loc[i, "R_Score"] = 2
    else:
        df.loc[i, "R_Score"] = 1
```

```python
df["R_Score"] = df["R_Score"].astype(int)
```

```python
# Copio direttamente NumOfProducts come F_Score e Balance come M_Score
df["F_Score"] = df["NumOfProducts"]
```

```python
df["M_Score"] = None

# Assegna M_Score=1 a chi ha Balance=0
df.loc[df["Balance"] == 0, "M_Score"] = 1

# Calcola qcut solo sui clienti con Balance > 0
mask = df["Balance"] > 0
df.loc[mask, "M_Score"] = pd.qcut(df.loc[mask, "Balance"], 4, labels=[2, 3, 4, 5]).astype(int)
```

```python
df.head()
```

```python
# Combina i tre punteggi in un codice RFM
df["RFM_Score"] = df["R_Score"].astype(str) + df["F_Score"].astype(str) + df["M_Score"].astype(str)
```

```python
# STEP 1 - scegliere feature e dividere il dataset per fase train/test

cols_to_use = [
    "CreditScore", "Age", "Tenure", "Balance",
    "NumOfProducts", "HasCrCard", "IsActiveMember", 
    "EstimatedSalary", "Recency"]

X = df[cols_to_use].copy() #copy imp perche crea una copia e non tocca il df
y = df["Exited"].astype(int)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size = 0.20,     #20% del dataset per test
    stratify = y,         # mantiente la stessa % di churn per le 2 parti
    random_state = 42     # riproducibile ad ogni lancio
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression


#Pipeline è un contenitore che mette in fila piu step
#scaler è lo step che trasforma i dati numerici in una scala standard mentre logreg è il modello


pipe = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("logreg", LogisticRegression(
        max_iter = 1000,
        class_weight = "balanced",
        random_state = 42
    ))
])


pipe.fit(X_train, y_train)

```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

y_pred = pipe.predict(X_test)  
y_proba = pipe.predict_proba(X_test)[:,1]

print(classification_report(y_test, y_pred, digits = 3))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("ROC AUC:", round(roc_auc_score(y_test, y_proba), 3))
```

con y_pred abbiamo effettuato le previsioni
prende i dati di X_test e fa predizioni
restituisce 0 o 1
e y_pred è un array di questi valori

y_proba restituisce le probabilità che il cliente sia in ciascuna classe es. [0.75, 0.25] vuol dire
75% probabilità di classe 0 (no churn)
25% probbilità di classe 1 (churn)
con [:,1] prenderà solo la seconda colonna
quindi y_proba conterrà solo la probabilità di churn per ogni cliente

successivamente ho stampato le metriche di controllo
Classification report confronta i valori reali con i valori predetti
e mostra  
PRECISION = quante volte il modello, quandi dice Churn, ha poi effettivamente ragione  
RECALL = quanti clienti churn reali riesce a trovare  
F! Score = media tra precision e recall  
ACCURACY = quante predizioni sono corrette in totale  

Confusion matrix 

[[ TN, FP],
[FN, TP]]

True Negative = clienti previsti come NON CHURN e lo sono davvero
False Positive = clienti previsti come Churn e non lo sono
False Negative = clienti previsti come non churn ma invece lo sono
True Positive = clienti previsti come churn e lo sono davvero

ROC_AUC = area sotto la curva ROC

1 il modello è perfetto
0.5 il modello prevede a caso

```python
#capiamo quali feature "pesano" di più per la previsione

coefs = pd.Series(
    pipe.named_steps["logreg"].coef_[0],      #andiamo a prendere da pipe i singoli step e prendiamo solo logreg per selezionare i coeff associati
    index = cols_to_use              #creiamo quindi una series i valori sono i coeff mentre l'index sono i nomi delle colonne (cols_to_use)
).sort_values(ascending = False)

print(coefs)
```

## Fase di clustering con K-means
preparazione delle feature per il clustering e scaler per standardizzare le scale

```python
features_for_clustering = [
    "CreditScore",
    "Age",
    "Balance",
    "Tenure",
    "NumOfProducts",
    "EstimatedSalary",
    "Recency" 
]
```

```python
X_cluster = df[features_for_clustering].copy()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)
```

scegliere il numero di cluster con Elbow method

```python
from sklearn.cluster import KMeans

inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Grafico del gomito
plt.plot(K_range, inertia, marker='o')
plt.xlabel('Numero di cluster (k)')
plt.ylabel('Inertia')
plt.title('Elbow method')
plt.show()
```

l'elbow (o gomito) è il punto in cui l'inerzia inizia a diminuire più lentamente e la pendenza della curva cambia
un numero di cluster ottimale pobrebbe essere 4 o 5

Asse Y = inertia → somma delle distanze quadratiche tra i punti e il loro centroide (minore = meglio)



```python
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
```
