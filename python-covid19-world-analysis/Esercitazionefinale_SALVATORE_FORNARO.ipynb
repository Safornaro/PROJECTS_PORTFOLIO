```python
# b. per ogni continente: la percentuale rispetto al totale mondiale del numero di casi

totale_casi_mondiale = df[df["location"] == "World"]["total_cases"].max()  #calcolo il totale casi mondiale

casi_per_continente = df.groupby("continent")["total_cases"].max()
casi_per_continente = casi_per_continente.reset_index()
casi_per_continente.columns = ["Continenti","Casi totali"] 
casi_per_continente = casi_per_continente.set_index("Continenti")

casi_per_continente 
for continente in casi_per_continente.index :
    percentuale = (casi_per_continente.loc[continente, "Casi totali"] / totale_casi_mondiale) * 100
    print(continente, round(percentuale, 2), "%")
```

5) Riguardo le nazioni di Italia, Germania, Francia e Spagna in tutto il 2021:
a. mostrare, in maniera grafica oppure numerica, la somma dei pazienti ospitalizzati per ognuna (colonna hosp_patients)
b. se ci sono dati nulli, con un breve commento scrivere se puÃ² essere possibile gestirli tramite sostituzione o meno


```python
filtro_nazioni5 = df["location"].isin(["Italy", "Germany", "France" , "Spain"])
filtro_anno5 = df["year"] == 2021

df_europe5 = df[filtro_nazioni5 & filtro_anno5]
somme_hosp = df_europe5.groupby("location")["hosp_patients"].sum()
```

La colonna hosp_patients presenta valori nulli esclusivamente per la Germania nel 2021.
Considerando la presenza della colonna hosp_patients_per_million e la colonna population avremmo potuto calcolare i valori di hosp_patients da quelli presenti in hosp_patients_per_million ma purtroppo anche in questa colonna per la nazione Germania nell'anno 2021 i valori sono tutti NaN.

A questo punto credo che  non sia possibile applicare una sostituzione attendibile. (in assenza di dati non posso utilizzare la media, lâ€™interpolazione o forward/backward fill)

Tuttavia, si potrebbe eventualmente fare riferimento alla colonna weekly_hosp_admissions per integrare l'analisi, pur trattandosi di una metrica differente poichÃ© comunica il flusso settimanale di ricoveri e non i pazienti ospedalizzati ogni giorno.

```python
df_europe5 = df_europe5.dropna(subset=["hosp_patients"])
somme_hosp = df_europe5.groupby("location")["hosp_patients"].sum()
#rimuovo quindi i null escludendo la Germania dal grafico

plt.figure(figsize=(8,6))
somme_hosp.plot(kind="bar", color="blue")

plt.title("Totale pazienti ospedalizzati in Francia, Italia e Spagna nel 2021")
plt.xlabel("Nazione")
plt.ylabel("Totale pazienti ospedalizzati (milioni)")
plt.grid(axis='y')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

I valori presenti in weekly_hosp_admissions comunicano informazioni diverse rispetto a hosp_patients non ha senso quindi mettere a confronto i due parametri.
Potrebbe perÃ² essere utile confrontare weekly_hosp_admissions nel 2021 per tutte e quattro nazioni.

```python
df_weekly = df[filtro_nazioni5 & filtro_anno5][["location", "date", "weekly_hosp_admissions"]].dropna()


df_weekly_unique = df_weekly.groupby(["location", "date"]).first().reset_index()  
totali = df_weekly_unique.groupby("location")["weekly_hosp_admissions"].sum()


plt.figure(figsize=(8,6))
totali.plot(kind="bar", color="orange")

plt.title("Totale ammissioni ospedaliere settimanali nel 2021 per Francia Germania Italia e Spagna")
plt.xlabel("Nazione")
plt.ylabel("Pazienti ammessi (somma annuale)")
plt.grid(axis='y')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

```

Sebbene la Germania non disponga di dati sulla colonna hosp_patients per il 2021, lâ€™analisi del flusso settimanale delle ammissioni ospedaliere (weekly_hosp_admissions) mostra che ha registrato oltre 1,8 milioni di ricoveri, un valore comparabile a quello degli altri principali paesi europei.

Questo potrebbe indicare che la pressione sul sistema ospedaliero tedesco non Ã¨ stata di grado inferiore rispetto a Francia, Italia e Spagna.
Inoltre potremmo chiaramente dedurre che lâ€™assenza di dati hosp_patients Ã¨ dovuta ad una mancanza di rilevazione o comunicazione del dato giornaliero e chiaramente non ad una reale assenza di ospedalizzazioni nell'anno 2021.
